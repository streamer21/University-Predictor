```
import pandas as pd
import seaborn as sns
import numpy as np
sns.set_theme()
```

```
data=pd.read_csv("Admission_Predict.csv")
```

```
data.head()
```

Lets do some visualization with seaborn

```
sns.distplot(data['GRE Score'],bins=15)
```

```
sns.distplot(data['TOEFL Score'],bins=15,color='k')
```

```
sns.countplot(data['University Rating'])
```

```
sns.countplot(data['Research'])
```

```
sns.distplot(data['CGPA'],bins=25,color='crimson')
```

```
#Lets compare two of the attributes(columns) together and visualize it
```

```
sns.jointplot(data=data,x='GRE Score',y='TOEFL Score',kind='hex',color='lightcoral')
```

```
sns.jointplot(data=data,x='GRE Score',y='CGPA',kind='hex',color='greenyellow')
```

```
#instead of individually comparing these visualization , lets compare all the columns together!!!!!
```

```
sns.pairplot(data,palette='winter')
```

```
#Now comes the most important part in data visualization. We will visualize how each of the features(columns) are correlated to each other
```

```
data=data.drop(['Serial No.'],axis=1)
```

```
tip=data.corr()
sns.heatmap(tip,cmap='YlGnBu')
```

```
#Now lets do some more visualization
```

```
data_new=data[(data['University Rating']==5) & (data['Chance of Admit ']>0.9)]
```

```
sns.distplot(data_new['GRE Score'],bins=7,color='darkcyan')
```

```
sns.distplot(data_new['TOEFL Score'],bins=5,color='darkorange')
```

```
sns.countplot(data_new['Research'],color='orchid')
```
